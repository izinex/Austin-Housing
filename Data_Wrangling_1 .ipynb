### 1.5 Fixing Datatypes of Columns<a id='1.5_Fixing_datatypes_of_columns'></a>

```python
print(pd.DataFrame(df.dtypes))
#All the data types seem in order there is no column which seems out of place expect for the object columns 
#lets look into all the columns which are object to see if we can change its type to a more approriate datatype
```

```python
obj_list = []
for i in df.columns:
    if df[i].dtype == 'O':
        obj_list.append(i)
    else:
        continue
```

```python
df[obj_list].head(10)
#Okay it looks like these are reseverd mostly for string values but lets look at latest_saledate we can turn this to
# datatime object
```

```python
df['latest_saledate'] = pd.to_datetime(df['latest_saledate'])
```

```python
#lets check all the int and float variables to see if we can turn them in values that consume less memory
num_list = []
for i in df.columns:
    if df[i].dtype == 'int64' or df[i].dtype == 'float64':
        num_list.append(i)
    else:
        continue
```

```python
df[num_list].head()
```

```python
df[num_list].describe()
# we can see we can turn some of the columns to smaller int type since they are low integers such as parkingSpaces,
# yearBuilt, numPriceChanges, numOfElementarySchools, numOfHighSchools, avgSchoolSize, MedianStudentsPerTeacher, numOfBathrooms
# numOfBedrooms and numOfStories
# We can also reduce the float size of some of the float types, like avgSchoolDistance and avgSchoolRating
```

```python
df['yearBuilt'] = df['yearBuilt'].astype('uint16')
df['garageSpaces'] = df['garageSpaces'].astype('uint8')
df['parkingSpaces'] = df['parkingSpaces'].astype('uint8')
df['numPriceChanges'] = df['numPriceChanges'].astype('uint8')
df['numOfElementarySchools'] = df['numOfElementarySchools'].astype('uint8')
df['numOfMiddleSchools'] = df['numOfMiddleSchools'].astype('uint8')
df['numOfHighSchools'] = df['numOfHighSchools'].astype('uint8')
df['avgSchoolDistance'] = df['avgSchoolDistance'].astype('float16')
df['avgSchoolRating'] = df['avgSchoolRating'].astype('float16')
df['latitude'] = df['latitude'].astype('float16')
df['longitude'] = df['longitude'].astype('float16')
df['avgSchoolSize'] = df['avgSchoolSize'].astype('uint16')
df['MedianStudentsPerTeacher'] = df['MedianStudentsPerTeacher'].astype('uint8')
df['numOfBathrooms'] = df['numOfBathrooms'].astype('uint8')
df['numOfBedrooms'] = df['numOfBedrooms'].astype('uint8')
df['numOfStories'] = df['numOfStories'].astype('uint8')

# I decided to leave zpid, zipcode, and propertyTaxrate as it is

```
